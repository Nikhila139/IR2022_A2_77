```python
#importing the necessary libraries such as pandas, numpy, matplotlib and math.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#reading the given file into a pandas dataframe.
df = pd.read_csv('IR-assignment-2-data.txt', sep=" ", header=None)
df
```

```python
#considering the queries that are having only qid=4
qdf=df[df[1]=="qid:4"]
#qdf=qdf.reset_index(drop=True)
qdf
```

```python
#max dcg is obtained when the relevance scores are arranged in descending order
mdf=qdf.sort_values(by=0, ascending=False)
mdf=mdf.reset_index(drop=True)
mdf
```

```python
# Q) Make a file rearranging the query-url pairs in order of max DCG
np.savetxt(r'maxDCG.txt', mdf.values, fmt='%s')
```

```python
# Q) State how many such files could be made.
l=list(mdf[0].unique())
cnt=1
for i in l:
    c=len(mdf[mdf[0]==i])
    print(f"{i} = {c}")
    cnt=cnt*(math.factorial(c))
print(f"no. of files with max DCG {cnt}")
```

```python
#computing nDCG with the following function compute_nDCG
def Compute_nDCG(n,d):
    d = np.asfarray(d)[:n]
    dcg = d[0] + np.sum(d[1:] / np.log2(np.arange(2, d.size + 1)))
    
    """dcg=0
    for i in range(1,n+1):
        dcg += (math.pow(2,(d[i-1]-1))/math.log2(i+1))"""
    return dcg
```

```python
dcgmax=Compute_nDCG(len(mdf),mdf[0])
dcgmax
```

```python
# (a) At 50
dcg50=Compute_nDCG(50,qdf[0])
idcg50=Compute_nDCG(50,mdf[0])
ndcg50=dcg50/idcg50
print(ndcg50)
```

```python
# (b) For the whole dataset
dcgwhole=Compute_nDCG(len(qdf),qdf[0])
idcgwhole=Compute_nDCG(len(mdf),mdf[0])
ndcgwhole=dcgwhole/idcgwhole
print(ndcgwhole)
```

```python
rdf=qdf[qdf[0]>0]
rdf
```

```python
rr=list(rdf[0])
totalr=len(rr)
totalr
```

```python
r75=list(qdf[76])
r=list(qdf[0])
s=[]
for i in r75:
    j=i.split(':')
    s.append(float(j[1]))
```

```python
sr= [j for i,j in sorted(zip(s,r),reverse=True)]
```

```python
ct=0
cr=0
pl=[]
rl=[]
for i in sr:
    ct= ct+1
    if i>0:
        cr+=1
    pl.append(cr/ct)
    rl.append(cr/totalr)
```

```python
#The following is the graph plotted for Precision vs Recall
plt.title('Precision-Recall Curve')
plt.plot(rl,pl)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```
